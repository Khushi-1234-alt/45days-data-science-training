```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = sns.load_dataset("tips")
df.head()
```

```python
# sns.lineplot(df["total_bill"])
# sns.scatterplot(df["total_bill"])
sns.scatterplot(x = "total_bill", y = "tip", data = df)
plt.title("Scatterplot")
plt.show()
```

```python
sns.histplot(df["total_bill"],color = "lime" , edgecolor = "black")
```

```python
sns.barplot(x = "day" , y = "total_bill"  , data = df)
```

```python
sns.kdeplot(df["total_bill"])
## kdeplot in Python refers to a Kernel Density Estimate plot, 
##  a visualization method used to represent the probability density function of continuous variables. 
```

```python
##sns ---> 1 categorical ---> catplot
## numerical ---> scatter plot ,lineplot ---> relational ---> relplot
## catplot 
tips = sns.load_dataset("tips")

# catplot: show boxplot of total_bill by day
# sns.catplot(x="day", y="total_bill", kind="box", data=tips)
# plt.show()

## relplot
# tips = sns.load_dataset("tips")

# # relplot: scatter plot
# sns.relplot(x="total_bill", y="tip", kind="scatter", data=tips)
# plt.show()

# relplot: line plot (example)
sns.relplot(x="total_bill", y="tip", kind="line", data=tips)
plt.show()
```

```python
sns.relplot(x = "total_bill",y = "tip",data = df,
            kind="scatter",
            hue="smoker",
            style="size",
            col="day")
```

```python
## catplot
sns.catplot(x = "day",y = "total_bill",data = df,
            kind="swarm")
```

```python
## count plot 
sns.countplot(x = "day" , data = df)
```

```python
## pointplot ---> shows the mean of the data connected by lines to show the trend
sns.pointplot(x = "day",y = "total_bill",data = df,hue="smoker")
```

```python
## boxplot ---> spread of data in the form of box
sns.boxplot(x = "day", y = "total_bill", data = df)
```

```python
## violinplot ---> spread of data in the form of violin
sns.violinplot(x = "day" , y = "total_bill" , data = df)
```

```python
##outliers handle 
q1 = df["total_bill"].quantile(0.25)  # q1 = 25th percentile (lower quartile)
q3 = df["total_bill"].quantile(0.75)  # q3 = 75th percentile (upper quartile)
iqr = q3-q1
lower_limit = q1-1.5*iqr   ##  Calculate lower & upper limits
upper_limit = q3+1.5*iqr
print(lower_limit,upper_limit)
## Check which rows are outliers
# df[df["total_bill"]<lower_limit]
df[df["total_bill"]>upper_limit]
## remove outliers
df = df[(df["total_bill"]>lower_limit) & (df["total_bill"]<upper_limit)]
sns.boxplot(df["total_bill"])
```
