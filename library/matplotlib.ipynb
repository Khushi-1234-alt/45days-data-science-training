# MATPLOTLIB
#### #matplot and seaborn -- data visualization-- graphs andd plots
#### lineplot , barplot , histogram , piechart , scatterplot

```python
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd 
import seaborn as sns 
```

```python
arr = np.random.randint(20000,40000,100)
arr
```

```python
## lineplot ---> numerical data 
arr.sort()
plt.plot(arr)
plt.show()
```

```python
x = [1,2,3,4,5]
y = [1,10,9,30,25]
plt.plot(x,y,marker = "s",ms = 10,
         mfc = "red",mec = "green",
         color = "black",
         lw = 3,ls = ":",
         alpha = 0.7,label = "lineplot")
plt.title("lineplot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.show()
```

```python
## scattrplot  
plt.scatter(x,y,marker="D")
plt.show()
```

```python
df = pd.read_csv("Used_Bikes.csv")
df.head()
```

```python
plt.scatter(df["kms_driven"],df["price"])
plt.plot(df["kms_driven"],df["price"])
```

```python
## bar graph
x = ["mp","up","rj","gj"]
y = [10,20,30,40]
colours = ["red","green","blue","yellow"]
hatch_patterns = ["//","\\\\","xx","oo"]
plt.barh(x,y)
plt.legend()
# plt.grid()
plt.show()
```

```python
## legend 
x = ["mp","up","rj","gj"]
y = [10,20,30,40]
colours = ["red","green","blue","yellow"]
hatch_patterns = ["//","\\\\","xx","oo"]
plt.bar(x,y,color = colours,hatch = hatch_patterns,
        label = x,width=0.3,edgecolor = "black",alpha = 0.7,
        linewidth = 3)
# plt.barh(x,y)
plt.legend()
plt.grid()
plt.show()
```

```python
## histogram
arr = np.random.randint(20000,40000,100)
plt.hist(arr , bins = 20)
plt.show()
```

```python
## piechart ---> percentage
x = ["mp","up","rj","gj"]
y = [10,20,30,40]
explodes  = (4,0,0,0)
plt.pie(y,labels = x,autopct="%1.2f%%",shadow=True,explode = explodes)
plt.savefig("piechart.pdf")
plt.show() 
```

```python
##subplot 
import warnings
warnings.filterwarnings("ignore")


x = np.linspace(0,10,100)
plt.figure(figsize=(10,6))
plt.subplot(2,2,1) ##2 rows , 2 columns , 1st graph
plt.plot(x,np.sin(x))
plt.title("sin")
plt.subplot(2,2,2) ##2 rows , 2 columns , 2nd graph
plt.plot(x,np.cos(x))
plt.title("cos")
plt.subplot(2,2,3) ##2 rows , 2 columns , 3rd graph
plt.plot(x,np.tan(x))
plt.title("tan")
plt.subplot(2,2,4) ##2 rows , 2 columns , 4th graph
plt.plot(x,np.log(x))
plt.title("log")
plt.tight_layout()
plt.show()
```
